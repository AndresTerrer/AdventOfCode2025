````python
import sys
sys.path.append("..")

from common.utils import Day, read_lines
test_path = "test.txt"
data_path = "data.txt"
````

````python
"""  
First, I need a data structure or a representation of the problem
lets read the lines with "^" and the starting line with "S".

If a beam hits a position i of a line, the next line will have incomming rays from positions
i-1 and i+1.

If the j+1 line does not have a splitter in any of those positions, the ray will continue to the next layer.

Store all rows into a list of characters (". ", "S" and "^")

keep an array of the same length as a row that will store the beams in this iteration.

ii terate over all layers, update current beams and count how many times the beam splits.

Judging by the data, I don't need to add failsafe because the index can not overflow according to the rules

"""
````

````python
class Day7(Day):
    def __init__(self, path):
        super().__init__(path)
        self.lines = [[char for char in line] for line in read_lines(path)]
        self.beams = [char == "S" for char in self.lines[0]]
        self.line_index = 1

    def propagate_beams(self) -> int:
        """  
        Propagate the current state of the beams throught the line
        indicated by self.line_index.

        Count the number of hits on this iterations and return it
        """
        
        hits = 0

        line = self.lines[self.line_index]
        has_a_splitter = [e == "^" for e in line]

        hits_a_splitter = [has and b for has,b in zip(has_a_splitter, self.beams)]
        hits = sum(hits_a_splitter)
        self.line_index = self.line_index + 1

        if hits == 0:
            return hits

        # Update the beams according to the rules
        hit_indeces = [i for i, x in enumerate(hits_a_splitter) if x == True]

        for hi in hit_indeces:
            new_beams = self.beams
            new_beams[hi] = False
            new_beams[hi-1] = True
            new_beams[hi+1] = True
        
        self.beams = new_beams
        return hits
    
    def solve_part_1(self):
        """
        loop and count the hits
        """
        result = 0
        for _ in range(len(self.lines) - 1):
            result += day7.propagate_beams()
        return result

    
day7 = Day7(test_path)
day7.solve_part_1()
````

````python
day7 = Day7(data_path)
day7.solve_part_1()
````

````python
"""
The number of timelines created in each layer is just 2Â·n_hits
we start with 1 timeline
"""

class Day7b(Day7):
    def __init__(self, path):
        super().__init__(path)
        self.lines = [[char for char in line] for line in read_lines(path)]
        self.beams = [char == "S" for char in self.lines[0]]
        self.line_index = 1
    
    def solve_part_2(self):
        result = 1
        for _ in range(len(self.lines) - 1):
            n_hits = super().propagate_beams()
            if n_hits == 0:
                continue
            result = result + 2 * n_hits
        return result


day7 = Day7b(test_path)
day7.solve_part_2()
````
